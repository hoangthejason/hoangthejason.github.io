# Logistic regression walkthrough

This notebook is rendered statically at build time.


```python
import math
x = 0.75
sigmoid = 1 / (1 + math.exp(-x))
print(f"sigmoid({x}) = {sigmoid:.4f}")
```

The decision rule is based on the probability estimate. If the probability is above a threshold such as $0.5$, predict the positive class.
